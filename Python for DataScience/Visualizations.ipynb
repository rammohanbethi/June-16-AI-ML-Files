```python
plt.figure(figsize=(12,5))
a=[1,2,3,4,5,6,7,8,9]
b=[7,7,7,7,7,7,7,7,7]
c=[7,8,9,10,7,11,12,13,14]

plt.plot(a,b,linewidth=4,label='Line-1 a&b')
plt.plot(a,c,linewidth=2,label='Line-2 a&c')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('A&B&C plot')
plt.legend()
plt.grid()
```

```python
x1=[1,3,5,7]
x2=[2,4,6,8]
y1=[7,7,7,7]
y2=[18,18,29,40]

plt.figure(figsize=(8,4))
ax=plt.axes()
plt.bar(x1,y1,label='First bar graph',color='#ffe169') 
plt.bar(x2,y2,label='Second bar graph',color='#cf69ff')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar Graph')
plt.legend()
```

```python
num1=np.array([1,3,5,7,9])
num2=np.array([2,4,6,8,10])
plt.bar(num1,num1**2,width=0.3,color='#fff769')
plt.bar(num2,num2**2,width=0.3,color='#f2b56b')
for x,y in zip(num1,num1**2):
    plt.text(x,y+1,'%d' % y)
for x,y in zip(num2,num2**2):
    plt.text(x,y+1,'%d' % y)
```

Donut, standard

```python
plt.figure(figsize=(8,8))
labels=['python','php','C','Java']
number=[75,70,85,60]
colors=['red','green','yellow','blue']
explode=[0.1,0,0,0.1]
plt.pie(number,labels=labels,autopct='%1.2f%%',colors=colors,explode=explode,shadow=True)
plt.legend()
```

```python
#Donut plot
plt.figure(figsize=(9,5))
area=[48,30,20,15]
labels=['low','medium','high','critical']
colors=['red','green','yellow','blue']
plt.pie(area,labels=labels,colors=colors)
my_circle=plt.Circle((0,0),0.7,color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
```
